### 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data.head(2)
```

```python
def rating(rating):
    if rating > 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
data['rating_category'] = data['Rating'].apply(rating)
```

### 22. Find Unique Values From Genre 

```python
data.head(5)
```

```python
list1 = []
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d = []
for item in list1:
    for i in item:
        one_d.append(i)
one_d
```

```python
uni_list = []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
